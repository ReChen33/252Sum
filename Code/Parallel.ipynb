2.1. Per-user process limit (ulimit -u)

ulimit -u
That number (for example, “1024”) is the maximum number of processes (including threads, since threads often count toward the process limit) you can spawn simultaneously.

If you hit that limit and try to fork() another process, you’ll see “Resource temporarily unavailable” errors.

To see all current limits for your shell session, run:

ulimit -a
Look for the line that says max user processes (or “max user processes (-u)”).
